```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/Table_2A_State_Uts.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.columns
```

```python
data=data.rename(columns={'India/State/Union Territory':'State','Population 2011':'Population', 'Decadal Population Growth Rate - 2001-2011':'Growth Rate','Population Density (per sq.km) - 2011':'Density'})
```

```python
data.head()
```

```python
data=data[data['State']!='India']
```

```python
data=data.drop_duplicates()
```

```python
data.isnull().sum()
```

```python
data['Growth Rate']=data['Growth Rate'].astype(str).str.replace('%','')
data['Growth Rate']=pd.to_numeric(data['Growth Rate'])
data['Population']=pd.to_numeric(data['Population'])
data['Density']=pd.to_numeric(data['Density'])
```

```python
data.head()
```

```python
data.describe()
```

```python
data.to_excel('Cleaned_population_dataset.xlsx',index=False)
```
